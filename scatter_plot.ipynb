```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_excel('NewOrders.xlsx')
```

```python
df.head(5)
df.columns
```

```python
X = df['Sales']
```

```python
y = df['Profit']
```

```python
sns.relplot(X,y,data=df)
```

```python
sns.relplot(X,y,data=df,hue='Order Priority')#hue='Order Priority' is a categorical variable
```

```python
sns.relplot(X,y,data=df,hue='Order Priority',style='Order Priority')# style gives different shapes
```

```python
sns.relplot(X,y,data=df,hue='Order Priority',style='Ship Mode')#Ship mode is a categoical variale
```

```python
sns.relplot(X,y,data=df,size='Discount') # Discount value determie the size of dots
```

```python
sns.relplot(X,y,data=df,size='Discount',sizes=(25,200))
```
