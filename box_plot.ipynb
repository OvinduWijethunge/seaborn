```python
import seaborn as sns
import pandas as pd
df = pd.read_excel('NewOrders.xlsx')
df.head(5)
```

```python
df.columns
```

```python
# box plot gives how your data distributed and mean ..quantile..
```

```python
sns.catplot(x='Product Category',y='Shipping Cost',data=df,kind='box')
```

```python
sns.catplot(x='Product Category',y='Shipping Cost',data=df,hue='Ship Mode',kind='box')
```

```python
sns.catplot(x='Product Category',y='Shipping Cost',data=df,hue='Ship Mode',hue_order=['Delivery Truck','Express Air','Regular Air'],kind='box')
```

```python
sns.catplot(x='Product Category',y='Shipping Cost',data=df,col='Ship Mode',kind='box')
# share same Y axis
```

```python
sns.catplot(x='Product Category',y='Shipping Cost',data=df,row='Ship Mode',kind='box')
#share same X axis
```

```python
sns.catplot(x='Product Category',y='Shipping Cost',data=df,col='Ship Mode',row='Customer Segment',kind='box')
```
