```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_excel('NewOrders.xlsx')
```

```python
df.head(5)
```

```python

df.columns
```

```python
sns.relplot('Order Date','Sales', data=df, kind='line')
```

```python
df['year'] = df['Order Date'].dt.year
```

```python
df['year'].unique()
```

```python
sns.relplot('year','Sales', data=df, kind='line')
```

```python
sns.relplot('year','Sales', data=df, kind='line',ci=None)
```

```python
sns.relplot('year','Sales', data=df, kind='line',ci='sd')
```

```python
sns.relplot('year','Sales', data=df, kind='line',estimator=None)
```

```python
sns.relplot('year','Sales', data=df, kind='line',estimator=sum)
```

```python
sns.relplot('year','Sales', data=df, kind='line',hue='Ship Mode')
```

```python
sns.relplot('year','Sales', data=df, kind='line',hue='Ship Mode',style='Product Category')
```

```python
sns.relplot('year','Sales', data=df, kind='line',hue='Ship Mode',style='Product Category',ci=None)
```

```python
sns.relplot('year','Sales', data=df, kind='line',hue='Ship Mode',col='Product Category')
```

```python
sns.relplot('year','Sales', data=df, kind='line',hue='Ship Mode',col='Region',ci=None)
```

```python
sns.relplot('year','Sales', data=df, kind='line',hue='Ship Mode',col='Region',ci=None,col_wrap=2)
```
